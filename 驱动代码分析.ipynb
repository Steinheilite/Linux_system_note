### ***看老妈写的ADC驱动*** ###
ADC analog to digital converter 
将模拟信号转换成数字
        在模拟信号中：自然信号，比如3.1  3.3 不一样
        在数字信号中：强制转换的数字信号  3.1 3.3可能代表一个内容
这个芯片将模拟信号转换成数字，取决于数据的精度。\

*接口函数和驱动程序*\

ADC模块驱动提供了四个不同的接口函数:
- 'adc_open()'初始化寄存器
- 'adc_close'
- 'adc.mmap' map a ring buffer 建立一个环形缓冲区并映射到用户空间
- 'adc.ioctl'\

ADC驱动程序:
- s2s65a_adc.c
- s2s65a_common_user.h结构体定义\\

**Sequence（时序图）**\
    初始化： 打开再做一次mmap\
    之后：启动内核定时器，
        每20ms，中断调用函数进入驱动，启用中断
        通过驱动程序检查一次（是否转换结束），如果有结果就存入buffer失败则返回0xFFFF
    应用程序访问：应用程序可以命令停止转换， 程序可以阅读index里面转换的数据。\
    结束：
    
mmap内存映射：将这个设备的可操作空间映射到用户进程里面的虚拟地址（这样用户可以阅读驱动带来的数据）

## 数值常数的表示 ##
0UL 表示 无符号长整型 0\
1UL 表示 无符号长整型 1\
如果不写UL后缀，系统默认为：int, 即，有符号整数。
- 1.数值常数有：整型常数、浮点常数；
- 2.只有数值常数才有后缀说明；
- 3.数值常数后缀不区分字母大小写。

整型常数的表示形式有：十进制形式、以0开头的八进制形式、以0x开头的十六进制形式，无二进制形式。\
- 整型常数默认是signed int的。
- 对整型常数进行类型转换的后缀只有：u或U(unsigned)、l或L(long)、u/U与l/L的组合(如：ul、lu、Lu等)。例：100u; -123u; 0x123l;\浮点常数的表示形式有：科学计数形式和小数点形式。

浮点常数默认是double的。
- 对浮点常数进行类型转换的后缀只有：f或F(单精度浮点数)、l或L(长双精度浮点数)。(注：因浮点型常数总是有符号的，故没有u或U后缀)。例：1.23e5f; 1.23l; -123.45f
**在这个程序里
0UL U :无符号， Llong: 长整型。加上~取反符号 根据的符号结合关系先取反再移位代指最后一位页号地址** 


\>>和<<运算符都是是移位操作。
- <<是将指定二进制向左移指定位数高位舍弃低位数字用零补齐
- \>>是向右移动特定位数左侧无符号数用零补齐低位舍弃
- & 按位与 如果两个相应的二进制位都为1，则该位的结果值为1，否则为0。
- l 按位或两个相应的二进制位中只要有一个为1，该位的结果值为1，否则为0

实际应用: 
- 清零，取反加上按位与就可以全变成0

（4）在定时器到期时，function被执行；

HZ: 时钟中断每秒触发的次数。

`ADC_RegSet8(g_p_adc_module, ADC_CONTROL_REG,BIT(5));`\
调用了8位寄存器的宏RegSet8 

`#define ADC_RegSet8(x, addr, data)	\
	( ADC_REGS8(x, (addr)) |= (data))`\

将地址合并之后转换为char 然后或等于data

data 写的是bit (5)  (1<<x)
1左移五位 

这样就把控制寄存器的第五位变成了1，其它位不受影响（原本是0还是0，1还是1）

同时控制寄存器每一位都有自己的意义，当第五位变成一之后进入工作模式
